**(b) (2 points)** Apply a 2D convolution and a median filter (kernel size of 11x11 pixels with equally distributed weights in both cases) to the grey image. Show the original, convolved and median filtered grey images using a 1x3 plot. Describe which of them is better to reduce the "salt and pepper" noise.

In this case, the better noise reduction of s&p would be 2d convolution, due to the mask selected for median filter involved more pixels values than expected

```python
from skimage.color import rgb2gray
from skimage import filters, util
import numpy as np
import skimage as ski # useful for image loading and saving
from matplotlib import pyplot as plt # useful for plotting images on iPython or Jupyter Notebooks
%matplotlib inline
from skimage import io
from skimage import data, exposure, img_as_float
from scipy import ndimage # Import library

noisyPears = io.imread('images/NoisyPears.png')

noisyPears_gray = rgb2gray(noisyPears)

mask_2d = np.ones((11,11))

mask_2d = mask_2d/np.sum(mask_2d)

convolved_pears_2d = ndimage.convolve(noisyPears_gray, mask_2d)

median_pears = noisyPears_gray.copy()

mask= np.ones((11,11))
mask = mask/np.sum(mask)

median_pears = filters.median(noisyPears_gray, mask)

representation, position = plt.subplots(ncols=3, nrows=1,figsize=(20,10))

position[0].imshow(noisyPears_gray, cmap = 'gray')
position[0].set_title('Original pears')
position[1].imshow(convolved_pears_2d, cmap = 'gray')
position[1].set_title('Convolted pears')
position[2].imshow(median_pears, cmap = 'gray')
position[2].set_title('Median pears')

plt.show()
```

```python
# Sinusoidal plot points generation
import numpy as np
import skimage
from skimage import io
import matplotlib.pyplot as plt

ic = io.ImageCollection('images/Barcelona-sequenceB/*.png')
        # Reading a sequence of images from a folder

import matplotlib.animation as animation
%matplotlib nbagg 
    #Changing the pluggin is necessary always when visualizing a video!

i=0       #Inicializing the video display
fig = plt.figure()  # Create figure
im = plt.imshow(ic[i], animated=True) #Visualize the first image

def updatefig1(i):   #Updating the frame visualization
    im.set_array(ic[i*5]) #Changing the content of the canvas
    
    return im, #to return a tuple!

ani = animation.FuncAnimation(fig, updatefig1, interval=5, blit=False, frames=50, repeat= False)

from skimage import *

# Histogram RGB
# Look for the difference between histograms with imagen It - It-1 [val absoluto] Media y diferencias
# Plot de las diferencias

def histogram_n_bins(image, bins):
    
    image_float = img_as_float(image) # Conversion image's dtype to float 
    
    red_channel, bin_centers = exposure.histogram(image_float[:,:,0], nbins = bins)
    green_channel, bin_centers = exposure.histogram(image_float[:,:,1], nbins = bins)
    blue_channel, bin_centers = exposure.histogram(image_float[:,:,2], nbins = bins)
    
    return (red_channel, green_channel, blue_channel)

%matplotlib nbagg 


hist_collection = list()

def data_gen():
    t = data_gen.t
    cnt = 0
    while cnt < 1000:
        hist = histogram_n_bins(ic[cnt], 8)
        rgb_mean = 0
        if len(hist_collection)!=0:
            red_difference = abs(hist[0] - hist_collection[len(hist_collection)-1][0])
            green_difference = abs(hist[1] - hist_collection[len(hist_collection)-1][1])
            blue_difference = abs(hist[2] - hist_collection[len(hist_collection)-1][2])
            
            rgb_difference = [red_difference, green_difference, blue_difference]
            rgb_mean = int(np.mean(rgb_difference))
        
        hist_collection.append((hist[0], hist[1], hist[2]))
        
        cnt+=1
        t += 0.05
        y = rgb_mean
        # y = np.cos(2*np.pi*t) * np.exp(-t/10.)
        # adapted the data generator to yield both sin and cos
        yield t, y

data_gen.t = 0
# create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(2,1)

# intialize a line object on the second axes for plotting
line, = ax2.plot([], [], lw=2, color='r')

ax2.set_ylim(-1, 15000)
ax2.set_xlim(0, 16)
ax2.grid()

# initialize the data arrays 
xdata, ydata = [], []

def run(data):
    # update the data plot
    t, y = data
    xdata.append(t) # time = x axis
    ydata.append(y) # y axis

    # Plot image on top row
    ax1.imshow(ic[len(xdata)])
          
    # Plot sin in bottom row
    xmin, xmax = ax2.get_xlim()
    if t >= xmax:
        ax2.set_xlim(xmin, 2*xmax)
        ax2.figure.canvas.draw()
            
    # update the data of both line objects
    line.set_data(xdata, ydata)
    line.set_data([xdata],[ydata])
    return line

ani = animation.FuncAnimation(fig, run, data_gen, blit=True, interval=10, repeat=False)
plt.show()
```

```python
#method to calculate the mean
def calcul_mitjana():
    ic = io.ImageCollection('images/Barcelona-sequenceB/*.png')

    red_a, blue_a, green_a = histogram_n_bins(ic[0],8) #we calculate the histogram for the first photo in ic collection

    funcio = []

    for fotograma in ic[1:]:#for all photos in ic except the first
        red, green, blue = histogram_n_bins(fotograma,8) #we calculate the histogram

        redT = np.linalg.norm(red-red_a) #we substract the previous image to the actual one 
        greenT = np.linalg.norm(green-green_a)
        blueT = np.linalg.norm(blue-blue_a)

        mitjana = (redT+greenT+blueT)/(3*8) #we  calculate the mean of the 3 channels

        funcio.append(mitjana)#and append this mean to the function we are creating
        
        #saving previous values
        red_a = red
        green_a = green
        blue_a = blue
    funcio[0] = 0
    return funcio


%matplotlib inline

llista_valors = calcul_mitjana()
plt.plot(llista_valors, "")
plt.show()
```
